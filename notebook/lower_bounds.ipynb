```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.special import gammaln, digamma
```

```python
plt.rcParams['font.size'] = 14
```

```python
x = np.linspace(0.1, 6, 1000)
x0, a = 3, 4
y1 = gammaln(x) - gammaln(x + a)
y10 = gammaln(x0) - gammaln(x0 + a)
y1_lower = y10 + (digamma(x0 + a) - digamma(x0)) * (x0 - x)
```

```python
y2 = gammaln(x + a) - gammaln(x)
y20 = gammaln(x0 + a) - gammaln(x0)
y2_lower = y20 + x0 * (digamma(x0 + a) - digamma(x0)) * (np.log(x) - np.log(x0))
```

```python
fig, axes = plt.subplots(1, 2, figsize=[10, 4])
fig.subplots_adjust(left=0.05, right=0.95, bottom=0.08, top=0.92)

ax = axes[0]
ax.plot(x, y1, c="black", linewidth=2)
ax.plot(x, y1_lower, c="r", linewidth=2)
ax.plot(x0, y10, "x", markersize=10, c="b")
ax.grid(which="major")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend([f"logΓ(x)-logΓ(x+{a})", f"lower bound using x0={x0}"])
ax.set_title('Equation(1)')

ax = axes[1]
ax.plot(x, y2, c="black", linewidth=2)
ax.plot(x, y2_lower, c="r", linewidth=2)
ax.plot(x0, y20, "x", markersize=10, c="b")
ax.grid(which="major")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend([f"logΓ(x+{a})-logΓ(x)", f"lower bound using x0={x0}"])
ax.set_title('Equation(2)')

fig.savefig("../lower_bound.png")
```
